# 3.1 2차원 그래프 그리기

### 리스트1 : 꺾은선 그래프 그리기

```python
# 리스트 1-(1)
import numpy as np
import matplotlib.pyplot as plt  # 그래프 그릴라면 matplotlib의 pyplot 이라는 라이브러리 가져와야함 => plt로
%matplotlib inline 
%matplotlib inline  
# 주피터 노트북에서 그래프를 표시하기 위해서

# data 작성
np.random.seed(1)  # 난수를 고정 => 랜덤 숫자를 만들기 전 처음 시작할 숫자. 진짜 랜덤이 아니라 특정 알고리즘에 의해 난수 생성되므로
                                   # 같은 시드값 넣어주면 같은 난수 생성됨. => 따라서 늘 같은 그래프 나옴
x=np.arange(10)
y=np.random.rand(10)  # *0~1사이*의 난수 10개 생성

#  그래프 표시
plt.plot(x,y)  # 꺾은선 그래프를 등록
plt.show()  # 그래프 그리기
```

### 리스트2 : 3차함수 f(x)그리기

```python
# 리스트 2-(1) : 3차함수 f(x) 그리기
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f(x):  # 함수 정의
    return (x-2)*x*(x+2)
```

```python
# 리스트 2-(2)
print(f(1))
```

```python
# 리스트 2-(3)
print(f(np.array([1,2,3])))  # x에 np.array([1,2,3])이라는 배열 대입 => ndarray로 돌려줌 => ,(콤마) 없음! (벡터꼴)
```

```python
# 리스트 2-(4) 
x=np.arange(-3, 3.5, 0.5)  #그래프를 그리는 x의 범위 : -3~3, x의 간격 : 0.5 => (-3,3,0.5)로 하면 2.5까지만 나오니까, 3.5로 해야함!
print(x)
```

### np.linspace(n1, n2, n) : 범위 n1에서 n2 사이를 일정간격의 n개 구간으로 나눠줌. (n1과 n2포함.. 총 10개 x)

```python
# 리스트 2-(5)
x=np.linspace(-3,3,10)  # 라인 아니고 lin space임!! x를 -3에서 3사이를 *일정간격*의 10개 구간으로 나눠줌. (lin.. 선형 공간...?)
print(np.round(x,2))   # x를 출력하는데, 소숫점 아래 둘째자리까지만 반올림해서 나타내기
```

```python
# 리스트 2-(6)
plt.plot(x, f(x))  # x축이 x, y축이 f(x)인 그래프 만들기
plt.show()
```

```python
# 리스트 2-(7)
# 함수를 정의

def f2(x,w):
    return (x-w)*x*(x+2)  # 함수 f2 정의

# x를 정의
x=np.linspace(-3,3,100)  # x를 100개 구간으로 분할.. -3과 3 포함해서 => 따라서 매끄러운 3차그래프 그려짐!

# 차트 묘사 : w가 달라지면(w=2, w=1) f2 그래프의 모양도 달라짐 => x축과 만나는 점들이 달라지므로
plt.plot(x, f2(x,2), color='black', label='$w=2$')  # color=(255,0,0) 처럼 tuple형태로 RGB값 지정 가능
plt.plot(x, f2(x,1), color='cornflowerblue', label='$w=1$')
plt.legend(loc="upper left")     # 범례 표시. 왼쪽 상단에 범례 표시하겠다. (loc=" "을 통해 범례 위치 설정)
plt.ylim(-15, 15)                # y축의 범위
plt.title('$f_2(x)$')            # 제목
plt.xlabel('$x$')                # x 라벨을 x축 옆에 표시하겠다
plt.ylabel('$y$')                # y 라벨을 y축 옆에 표시하겠다
plt.grid(True)                  # 그리드 그리겠다!(True)
plt.show()
```

### 여러개의 그래프 그리기 : plt.subplot(n1, n2, n) => 세로 n1, 가로 n2로 나눈 n번째에 그려짐. n은 1부터!!(0아님)

```python
# 리스트 2-(9)
plt.figure(figsize=(10,3))  # figure 지정 => 전체 영역의 크기 지정. (w,h). 각 그리드의(그래프의) 가로 길이는 10, 세로 길이는 3
plt.subplots_adjust(wspace=0.5, hspace=0.5)  # 그래프의 간격 지정
for i in range(6):
    plt.subplot(2,3,i+1)  # 그래프 묘사의 *위치* 지정. i는 0부터 시작하므로, subplot의 인자엔 i+1해서 1번째로 시작하도록 해야함!
    plt.title(i+1)
    plt.plot(x, f2(x,i), 'k')  # 그래프 그리기 ==> color= 생략 가능. cmyk의 k. => 1번 그래프엔 i=0이 고정으로 들어간 채로 x값 100개로 ㄱㄱ
    plt.ylim(-20, 20)  #y의 범위를 그래프들 모두 일정하게 보이기 위해서.. 이 코드 안쓰면 y의 범위 다 다르게 나옴
    plt.grid(True)
plt.show()
```

# 3.2 3차원 그래프 그리기 : 이변수 함수

```python
# 리스트 3-(1)
import numpy as np
import matplotlib.pyplot as plt

# 함수 f3 정의
def f3(x0, x1):
    r=2*x0**2 + x1**2  # **은 제곱
    ans = r*np.exp(-r)  # np.exp(x) : 지수함수
    return ans

# x0, x1에서 각각 f3을 계산
xn=9
x0=np.linspace(-2, 2, xn)  # 계산할 x0의 범위 설정. xn=9이므로, x0과 x1은 9개의 값으로 구성됨.
x1=np.linspace(-2, 2, xn)  
y=np.zeros((len(x0), len(x1)))  # 계산결과를 넣을 2차원 배열 변수 y 준비. 
for i0 in range(xn):     # xn=9. 따라서 i0과 i1은 0~8까지의 수. 총 9번 반복.
    for i1 in range(xn):
        y[i1, i0]=f3(x0[i0], x1[i1])  # 계산결과(ans)를 넣을 2차원 배열 변수 y 준비.f3(0,0)~f3(8,8)까지 2차원 배열 y[i0, i1]에 값 담음.
                                      # [i1, i0] 순서에 주의!! 
```

```python
# 리스트 3-(2)
print(x0)
```

```python
# 리스트 3-(3)
print(np.round(y,1))  # 걍 y로 2차원 배열(행렬) 출력할 수 있지만, np함수인 round를 사용해 y를 소수점아래 n자리로 반올림

# 결과: 중심과 주변이 0인, 도너츠 같은 모습
```

### 수치를 색으로 표현하기 : plt.pcolor(y) => y는 2차원 ndarray형 배열(행렬, 벡터)

```python
# 리스트 3-(4)
plt.figure(figsize=(3.5, 3))  # 그래프의 가로 : 3.5, 세로 : 3 크기
plt.gray()  # 흑백으로 표현 (회색 음영으로 표현). plt.jet(), plt.pink(), plt.bone()등 다양한 그라데이션 패턴 있음.
plt.pcolor(y)  # y : 2차원 배열 즉, 행렬 y를 색상으로 표현하겠다. (원래대로면 y는 위와같이 숫자로 구성된 행렬로 나옴). 우측에 컬러바ㅇ.
plt.colorbar()  # 우측에 컬러바 표시하겠다.
plt.show()
```

### ※함수의 표면을 표시하기 : 서피스 => ax.plot_surface()

```python
# 리스트 3-(5)
from mpl_toolkits.mplot3d import Axes3D  # 3차원으로 표시하기 위해 mpl_toolkits.mplot3d의(from) Axes3D import

xx0, xx1= np.meshgrid(x0, x1)  # 좌표점 x0, x1으로 xx0, xx1을 만들기

plt.figure(figsize=(5, 3.5))
ax=plt.subplot(1, 1, 1, projection='3d')  # 
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')  #
ax.set_zticks((0, 0.2))  #
ax.view_init(75, -95)  #
plt.show()
```

```python
# 리스트 3-(7)
print(xx0)  # np.meshgrid(x0, x1)으로 생성된 xx0은 2차원 배열이 됨.
```

### ※등고선으로 표시하기 : plt. contour()

```python
# 리스트 3-(9)
xn=50
x0=np.linspace(-2,2,xn)
x1=np.linspace(-2,2,xn)

y=np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0, x1)  #

plt.figure(1, figsize=(4,4))  # 1??
cont=plt.contour(xx0, xx1, y, 5, colors='black')  #
cont.clabel(fmt='%3.2f', fontsize=14)  #
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
