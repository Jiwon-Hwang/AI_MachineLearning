```python
# 리스트 5-1-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# 데이터 생성 --------------------------------
np.random.seed(seed=1) # 난수를 고정
X_min = 4 # X의 하한(표시 용) : X는 나이(최소 나이 : 4살)
X_max = 30 # X의 상한(표시 용)  (최대 나이 : 30살)
X_n = 16 # X의 상한(표시 용)
X = 5 + 25 * np.random.rand(X_n)
Prm_c = [170, 108, 0.2] # 생성 매개 변수
T = Prm_c[0] - Prm_c[1] * np.exp(-Prm_c[2] * X) \
+ 4 * np.random.randn(X_n) # (A)
np.savez('ch5_data.npz', X=X, X_min=X_min, X_max=X_max, X_n=X_n, T=T) # (B)
```

```python
# 리스트 5-1-(5)
# 데이터 그래프 ------------------------------
plt.figure(figsize=(4, 4))
plt.plot(X, T, marker='o', linestyle='None',
         markeredgecolor='black', color='cornflowerblue')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

```python
# 리스트 5-1-(6)  ==> 1017 실습 시작! / 저번시간이어서.. / 어떻게 3차원 그래프가 만들어지는지?!(meshgrid()이용)
from mpl_toolkits.mplot3d import Axes3D
# 평균 오차 함수 ------------------------------
def mse_line(x, t, w):  # "평균오차함수"!! ==> 입력값에 따른 y값 & 목적값 t 와의 차이점 에다가 제곱(자승) 취하고 평균 구해줌
    y = w[0] * x + w[1]  # 선형 회귀 모델, x,t,y 모두 벡터!
    mse = np.mean((y - t)**2)  # mean 안에 들어간 값도 벡터 /벡터의 원소별로 차를 취하고 자승/ mean : 다더해서 n분의1로 나눠주는..(평균내줌)
    return mse

# 계산 --------------------------------------
xn = 100 # 등고선 표시 해상도
w0_range = [-25, 25]  # w0 : 나이..? 서서히 변함 (아래 그래프 보면 알 수 있다.)
w1_range = [120, 170] # w1 : 완만하게 변함

x0 = np.linspace(w0_range[0], w0_range[1], xn)
x1 = np.linspace(w1_range[0], w1_range[1], xn)
xx0, xx1 = np.meshgrid(x0, x1)

J = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))

# 표시 --------------------------------------
plt.figure(figsize=(9.5, 4))
plt.subplots_adjust(wspace=0.5)

ax = plt.subplot(1, 2, 1, projection='3d')
ax.plot_surface(xx0, xx1, J, rstride=10, cstride=10, alpha=0.3,
                color='blue', edgecolor='black')
ax.set_xticks([-20, 0, 20])
ax.set_yticks([120, 140, 160])
ax.view_init(20, -60)

plt.subplot(1, 2, 2)
cont = plt.contour(xx0, xx1, J, 30, colors='black',
                   levels=[100, 1000, 10000, 100000])
cont.clabel(fmt='%1.0f', fontsize=8)
plt.grid(True)
plt.show()
```

```python
# 리스트 5-1-(7)
# 평균 제곱 오차의 기울기 ------------------------
def dmse_line(x, t, w):  # 목적함수를 w0과 w1으로 편미분한 값들 반환해주는 함수 정의! (for 나중에 경사하강법 적용하기 위해서!)
    y = w[0] * x + w[1]
    d_w0 = 2*np.mean((y-t)*x)  # w0에 대해서 편미분한 값 / x : 입력 벡터 / ***곱과 dot(내적)을 구분할 것!!!*** / 곱은 원소별로 곱하기
    d_w1 = 2*np.mean(y-t)    # w1에 대해서 편미분한 값
    
# d_w0: 목적 함수에 대한 w0의 미분값, d_w1:목적 함수에 대한 w01의 미분값
    return d_w0, d_w1  # 두 편미분한 결과 값을 반환! (따라서 이 함수는 편미분한 값들을 반환해주는 함수)
```

```python
# 리스트 5-1-(8)

# w0= 10, w1=165로 초기화
d_w=dmse_line(X,T, [10,165]) # w(가중치)가 10, 165로 주어짐
print(np.round(d_w, 1))
```

```python
# 리스트 5-1-(9)  ==> *경사하강법~! (코드 다시 체크해보기...결과그래프 모양 조금 다름)
# 구배법(경사하강법) ------------------------------------
def fit_line_num(x, t): # x : 입력벡터, t : 목적 백터(16 x 1인 세로 벡터)
    w_init = [10.0, 165.0] # 초기 매개 변수 ==> w0와 w1을 임의의 값으로 초기화하겠다.(10과 165로 초기화)
    alpha = 0.001 # 학습률
    i_max = 100000 # 반복의 최대 수 ==> 몇 번 반복시킬건지 정하기
    eps = 0.1 # 반복 종료 기울기의 절대 값의 한계 ==> 0에 가까워지면 최적값에 가까워졌다!(*최적점에 가까워지면 기울기 0에 가까워지므로)
    w_i = np.zeros([i_max, 2])  # weight 값이 어떻게 변해가는지 행렬에다가 저장(우선 0로 채워서 초기화해둠)
                                # np.zeros() 에는 [ , ] 이렇게 **[벡터 형식]**으로 넘기는 것 유의하기!!!
    w_i[0, :] = w_init  # 첫번째 행을 초기화한 값으로 할당!
    
    # w_i 업데이트     
    for i in range(1, i_max):  # for문 10000번 돌리기
        dmse=dmse_line(x, t, w_i[i-1])  # 미분한 값이 여기로 반환됨
        # 이 부분이 w 업데이트 하는 부분(계속 목적함수값 작아지는 방향으로...) 
        w_i[i,0] = w_i[i-1,0] - alpha*dmse[0]  # 새롭게 업데이트된 weight값! (최저점으로 가는, 함수 작아지는 방향으로 업데이트 됨)
        w_i[i,1] = w_i[i-1,1] - alpha*dmse[1]
        if max(np.absolute(dmse)) < eps:  # dmse의 두 원소들 모두에 절대값 취해서 큰 값을 취함 / 최저점에 도달시(?) for문 빠져나감!
            break   # eps보다 작아지지 않으면 계속 무한반복... 작아지면 break로 for문 빠져나옴
    w0 = w_i[i,0]  # 0번째 열 / for문 빠져나간 후에 업데이트 된 가중치 값 배열에 저장 ==> 이제 최종!
    w1 = w_i[i,1]  # 1번째 열
    w_1= w_i[:i, :]  # **0번부터 i번째 행까지만 행렬을 넘기겠다!!** & 열은 ():() 이므로 다 넘김ㅋ 
    return w0, w1, dmse, w_i


# 메인 ------------------------------------  < y = w0*x0 + w1 > : y=t, x0=나이

# ***1. 우선 그래프 그리기! (위에서 두 그래프 중 두번째 그래프)
plt.figure(figsize=(4, 4)) # MSE의 등고선 표시
xn = 100 # 등고선 해상도
w0_range = [-25, 25]
w1_range = [120, 170]
x0 = np.linspace(w0_range[0], w0_range[1], xn)
x1 = np.linspace(w1_range[0], w1_range[1], xn)
xx0, xx1 = np.meshgrid(x0, x1)
J = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))
cont = plt.contour(xx0, xx1, J, 30, colors='black',
                   levels=(100, 1000, 10000, 100000))
cont.clabel(fmt='%1.0f', fontsize=8) 
plt.grid(True)

# ***2. 이제 경사하강법 적용하기!
# 구배법 호출
W0, W1, dMSE, W_history = fit_line_num(X, T)  #  W_history : w의 누적값들(히스토리)


# ***3.결과보기
print('반복 횟수 {0}'.format(W_history.shape[0]))  #  W_history에 저장되어있는 행의 개수 
print('W=[{0:.6f}, {1:.6f}]'.format(W0, W1))
print('dMSE=[{0:.6f}, {1:.6f}]'.format(dMSE[0], dMSE[1]))  # 마지막으로 편미분한 값 (0에 가까울 수 밖에 없음)
print('MSE={0:.6f}'.format(mse_line(X, T, [W0, W1])))  # 

# 이 부분에 의해서 어떻게 변하는지(수렴해가는지)에 대해서 그려지는 것임!! ==> 최종 결과 그래프는 이코드에 의해..
# 가운데에 파란색 막대에서 가장 아래쪽부분이 최저점(최적점)
plt.plot(W_history[:, 0], W_history[:, 1], '.-',
         color='gray', markersize=10, markeredgecolor='cornflowerblue')
plt.show()
```

```python
# 리스트 5-1-(10)  ==> 위에서 w0과 w1은 구했음. 이제 이 그래프만 있으면(직선 모델), 예측 가능!!(ex. 나이에 따른 몸무게or키 예측기!)
#                  ==> 회귀문제 해결 가능!!

# 선 표시 ----------------------------------  # 직선을 그리는 함수
def show_line(w):
    xb = np.linspace(X_min, X_max, 100)
    y = w[0] * xb + w[1]
#선 그리기 
    plt.plot(xb, y, color=(.5, .5, .5), linewidth=4)



# 메인 ------------------------------------
plt.figure(figsize=(4, 4))
W=np.array([W0, W1])
mse = mse_line(X, T, W)
print("w0={0:.3f}, w1={1:.3f}".format(W0, W1))

# mse = mse_line(X, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
show_line(W) 
plt.plot(X, T, marker='o', linestyle='None',
         color='cornflowerblue', markeredgecolor='black') 
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()

# cf. 결과를 보면 훈련 데이터에 대해서 어느정도 fitting 했다고 보여짐
```

```python
# 리스트 5-1-(11)
# 해석해 ------------------------------------
def fit_line(x, t):
    mx = np.mean(x)
    mt = np.mean(t)
    mtx = np.mean(t * x)
    mxx = np.mean(x * x)
    w0 = (mtx - mt * mx) / (mxx - mx**2)  # 입력데이터(훈련데이터)와 목적벡터 주어지면 구할 수 있는 값.
    w1 = mt - w0 * mx
    return np.array([w0, w1])

# 메인 ------------------------------------
W = fit_line(X, T)
print("w0={0:.3f}, w1={1:.3f}".format(W[0], W[1]))
mse = mse_line(X, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.figure(figsize=(4, 4))
show_line(W)
plt.plot(X, T, marker='o', linestyle='None',
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

```python
# 리스트 5-1-(14)  ==> 나이와 몸무게 정해지면 키 정해짐
# 2차원 데이터의 표시 ------------------------
def show_data2(ax, x0, x1, t):  # 3차원... 막대 그래프 그려내는 함수
    for i in range(len(x0)):
        ax.plot([x0[i], x0[i]], [x1[i], x1[i]], 120, t[i], color='gray')
    # Z축에 수직인 3차원 라인을 그리려 Z값: 120~t[i]         
        ax.plot(x0, x1, t, 'o',color='cornflowerblue', markeredgecolor='black', markersize=6, markeredgewidth=0.5)
        ax.view_init(elev=35, azim=-75)

# 메인 ------------------------------------
plt.figure(figsize=(6, 5))
ax = plt.subplot(1,1,1,projection='3d')
show_data2(ax, X0, X1, T)
plt.show()
```

```python
# 리스트 5-1-(15)    
#면의 표시 ----------------------------------  < y=w0*x0 + w1*x1 + w3 > : 3차원 평면
def show_plane(ax, w):    
    px0 = np.linspace(X0_min, X0_max, 5)
    px1 = np.linspace(X1_min, X1_max, 5)
    px0, px1 = np.meshgrid(px0, px1)
    y = w[0]*px0 + w[1] * px1 + w[2]
    ax.plot_surface(px0, px1, y, rstride=1, cstride=1, alpha=0.3,
                    color='blue', edgecolor='black') 

#면의 MSE 함수 구현-----------------------------------
def mse_plane(x0, x1, t, w):
    y = w[0]*x0 + w[1]*x1 + w[2] # 
    mse=np.mean((y-t)**2)
    return mse

# 메인 ------------------------------------
plt.figure(figsize=(6, 5))
ax = plt.subplot(1, 1, 1, projection='3d')
W = [1.5, 1, 90]
show_plane(ax, W) 
show_data2(ax, X0, X1, T)  #  수직방향의 막대 그래프 그리는 부분! (지금은 왜 안그려지지ㅠㅠㅠ)
mse = mse_plane(X0, X1, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.show()
```

```python
# 리스트 5-1-(16)
# 해석해 ------------------------------------
def fit_plane(x0, x1, t):  
    c_tx0 = np.mean(t * x0) - np.mean(t) * np.mean(x0)
    c_tx1 = np.mean(t * x1) - np.mean(t) * np.mean(x1)
    c_x0x1 = np.mean(x0 * x1) - np.mean(x0) * np.mean(x1)
    v_x0 = np.var(x0)
    v_x1 = np.var(x1)
    w0 = (c_tx1 * c_x0x1 - v_x1 * c_tx0) / (c_x0x1**2 - v_x0 * v_x1)
    w1 = (c_tx0 * c_x0x1 - v_x0 * c_tx1) / (c_x0x1**2 - v_x0 * v_x1)
    w2 = -w0 * np.mean(x0) - w1 * np.mean(x1) + np.mean(t)
    return np.array([w0, w1, w2]) 

# 평편의 기울기 함수를 구현하여라.  ==> 시험문제가 이런식으로..?
# def dmse_plane(x0, x1, t, w):

# return d_w0,d_w1,d_w2

# 경사하강법 해 ------------------------------------
def fit_plane_num(x0,x1,t):
    w_init = [10.0, 10.0, 100.0] # 초기 매개 변수 
    alpha = 0.0001 # 학습률
    i_max = 100000 # 반복의 최대 수
    eps = 0.02 # 반복을 종료 기울기의 절대 값의 한계
    w = w_init

    return np.array([w[0], w[1], w[2]])


# 메인 ------------------------------------
plt.figure(figsize=(6, 5))
ax = plt.subplot(1, 1, 1, projection='3d')
W = fit_plane_num(X0, X1, T)
print("w0={0:.1f}, w1={1:.1f}, w2={2:.1f}".format(W[0], W[1], W[2]))
show_plane(ax, W)
show_data2(ax, X0, X1, T)
mse = mse_plane(X0, X1, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.show()
```
