```python
# 리스트 5-1-(6) : w와 J(mse)관계 / 어떻게 3차원 그래프가 만들어지는지?!(meshgrid()이용) => **목표 : 직방에서 최적의 w0과 w1을 구하기!
from mpl_toolkits.mplot3d import Axes3D
# 평균 오차 함수(mse) ------------------------
def mse_line(x, t, w):  # "평균오차함수"!! ==> 입력값에 따른 y값 & 목적값 t 와의 차이점 에다가 제곱(자승) 취하고 평균 구해줌
    y = w[0] * x + w[1]  # 선형 회귀 모델(*직선의 방정식), x,t,y 모두 ***벡터***(배열,행렬)
    mse = np.mean((y - t)**2)  # mean 안에 들어간 값도 벡터 /벡터의 원소별로 차를 취하고 자승/ mean :평균내줌(다 더해서 n분의1로 나눔)
    return mse  # *J (함수) 반환(평균 제곱 오차) / 두 매개변수(w[0], w[1]에 따른 J함수)

# 계산 --------------------------------------
xn = 100 # 등고선 표시 해상도(범위를 몇개의 구간으로 나눠서 축에 표시할지..)
w0_range = [-25, 25]  # w0_range: 배열. 0번지 값=-25, 1번지 값=25  (뒤에서 결국 처음과 끝값, 즉 범위로 쓰임) (이차함수 꼴)
w1_range = [120, 170] # w1 : 완만하게 변함 (왼쪽 벽쪽보면..아래로 완만하게 내려가는 형태)

x0 = np.linspace(w0_range[0], w0_range[1], xn) # 배열! -25~25사이를 xn만큼 등간격으로 쪼개서 나타내기..해상도! (이차함수꼴)
x1 = np.linspace(w1_range[0], w1_range[1], xn) # 배열! 120~170사이를 ...  (완만하게 변함. 직선 꼴)
xx0, xx1 = np.meshgrid(x0, x1)  # 좌표점 x0, x1으로 xx0 xx1 공간! 을 만들기 (3차원 ==> 직선꼴 변화 + 이차함수꼴 변화)

J = np.zeros((len(x0), len(x1)))  # len(x0) x len(x1) 차원의 행렬을 모두 0으로 초기화, J 선언 / xn x xn 개의 0
for i0 in range(xn):  # xn= 샘플의 개수(w0과 w1에 따라 16번*16번 for문 돌리고 J값(오차제곱평균) 구해놓고 ==> 각각을 2차원 행렬에 담기)
    for i1 in range(xn):
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))  # *** J : mse(평균제곱오차 목적함수) ***

# 표시 --------------------------------------
plt.figure(figsize=(9.5, 4))
plt.subplots_adjust(wspace=0.5)  # 각 개체(subplot들) 사이의 간격을 한 그래프의 가로길이의 절반으로!(9.5의 절반)

ax = plt.subplot(1, 2, 1, projection='3d')  # 1. 왼쪽 subplot (1x2짜리 행렬 중 첫번째 plot을 만들겠다!) ==> 그걸 ax라는 변수에 담기
ax.plot_surface(xx0, xx1, J, rstride=10, cstride=10, alpha=0.3,
                color='blue', edgecolor='black')  #  가로(row), 세로(column)로 몇 개의 선(뼈대)을 그을지 지정
ax.set_xticks([-20, 0, 20])  # 눈금 설정 (축부분에 눈금을 이렇게 3개만 표시하겠다)
ax.set_yticks([120, 140, 160])
ax.view_init(20, -60)  # 보는 각도(상하, 좌우)

plt.subplot(1, 2, 2)  # 2. 오른쪽 subplot (1x2짜리 행렬 중 두번째 plot을 만들겠다!)
cont = plt.contour(xx0, xx1, J, 30, colors='black',   
                   levels=[100, 1000, 10000, 100000])   # 등고선 plot 만들기() ==> for. 함수의 높이 알아보기
cont.clabel(fmt='%1.0f', fontsize=8)   # 각 등고선라인에 라벨 달기(숫자)
plt.grid(True)
plt.show()  # .plt들 다 보여주기 (1. 평균제곱오차 (함수)J 그래프 / 2. 평균제곱오차 J의 등고선 표시 그래프)
# 따라서 아래 두번째 등고선 그래프에서, 100이 J의 높이가 가장 낮은 곳이므로, 저 O 모양 안쪽 근처에서 최적값 가질 것으로 예측 가능!
```

```python
# 리스트 5-1-(7)  : 기울기를 계산하는 함수
# 평균 제곱 오차의 기울기 (J를 미분한 것. 그레디언트) ------------------------
def dmse_line(x, t, w):  # J의 그레디언트(d..델타..미분..)
    y = w[0] * x + w[1]
# d_w0: 목적 함수에 대한 w0의 미분값, d_w1:목적 함수에 대한 w1의 미분값
    d_w0 = 2 * np.mean((y - t) * x)  # y, t, x 모두 벡터!!(16개의 샘플의 나이, 키 값 갖고있는..)
    d_w1 = 2 * np.mean(y - t)
    return d_w0, d_w1
```

```python
# 리스트 5-1-(9)
# 경사 하강법 (***중요***) ----------------------------------
def fit_line_num(x, t):
    w_init = [10.0, 165.0] # 초기 매개 변수(임의의 w0, w1)
    alpha = 0.001 # 학습률
    i_max = 100000 # 반복의 최대 수
    eps = 0.1 # 반복을 종료 기울기의 절대 값의 한계
    w_i = np.zeros([i_max, 2])
    w_i[0, :] = w_init
    for i in range(1, i_max):
        dmse = dmse_line(x, t, w_i[i - 1])
        w_i[i, 0] = w_i[i - 1, 0] - alpha * dmse[0]
        w_i[i, 1] = w_i[i - 1, 1] - alpha * dmse[1]
        if max(np.absolute(dmse)) < eps: # 종료판정, np.absolute는 절대치
            break
    w0 = w_i[i, 0]
    w1 = w_i[i, 1]
    w_i = w_i[:i, :]
    return w0, w1, dmse, w_i


# 메인 ------------------------------------
plt.figure(figsize=(4, 4)) # MSE의 등고선 표시
xn = 100 # 등고선 해상도
w0_range = [-25, 25]
w1_range = [120, 170]
x0 = np.linspace(w0_range[0], w0_range[1], xn)
x1 = np.linspace(w1_range[0], w1_range[1], xn)
xx0, xx1 = np.meshgrid(x0, x1)
J = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        J[i1, i0] = mse_line(X, T, (x0[i0], x1[i1]))
cont = plt.contour(xx0, xx1, J, 30, colors='black',
                   levels=(100, 1000, 10000, 100000))
cont.clabel(fmt='%1.0f', fontsize=8) 
plt.grid(True)

# 구배법 호출
W0, W1, dMSE, W_history = fit_line_num(X, T)

# 결과보기
print('반복 횟수 {0}'.format(W_history.shape[0]))
print('W=[{0:.6f}, {1:.6f}]'.format(W0, W1))
print('dMSE=[{0:.6f}, {1:.6f}]'.format(dMSE[0], dMSE[1]))
print('MSE={0:.6f}'.format(mse_line(X, T, [W0, W1])))
plt.plot(W_history[:, 0], W_history[:, 1], '.-',
         color='gray', markersize=10, markeredgecolor='cornflowerblue')
plt.show()
# 그래프에서, 맨위의 점이 시작지점(w_init:초기값) / 맨아래가 종료지점(w)
```

```python
# 리스트 5-1-(11)
# 해석해 ------------------------------------
def fit_line1(x, t):
    mx = np.mean(x)
    mt = np.mean(t)
    mtx = np.mean(t * x)
    mxx = np.mean(x * x)
    w0 = (mtx - mt * mx) / (mxx - mx**2)
    w1 = mt - w0 * mx
    return np.array([w0, w1])

# 선형 회귀 모델의 해석해

def fit_line2(x, t):
    X = np.ones([len(x),2])
    T = np.zeros([len(t),1])
    for i in range(len(x)):
        T[i]=t[i]
        X[i,0] = x[i]
    
    X_T = np.transpose(X)
    XTX = X_T.dot(X)    
    XTX_inv =np.linalg.inv(XTX)
    XXX = XTX_inv.dot(X_T)
    w = XXX.dot(T)
    return [w[0,0], w[1,0]]

# 메인 ------------------------------------
W = fit_line1(X, T)
print(W)
print("w0={0:.3f}, w1={1:.3f}".format(W[0],W[1]))
mse = mse_line(X, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.figure(figsize=(4, 4))
show_line(W)
plt.plot(X, T, marker='o', linestyle='None',
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

### 1. 해석해(분석적 방법) : 목적함수 미분한 것(g) = 0이 되는 최적해 w을 구하는 문제(직접 방정식을 구하는 방법)
### 2. 경사하강법(수치적 방법) : 계속 update해서 최적해 w로 점점 접근해가는 문제(그레디언트가 0에 수렴하면 stop)          ==> 간단하지 않은 현실적인 목적함수..의 경우에.. 대부분이 수치적 방법으로 해결!

```python
# 리스트 5-1-(16)
# 해석해 ------------------------------------
def fit_plane1(x0, x1, t):
    c_tx0 = np.mean(t * x0) - np.mean(t) * np.mean(x0)
    c_tx1 = np.mean(t * x1) - np.mean(t) * np.mean(x1)
    c_x0x1 = np.mean(x0 * x1) - np.mean(x0) * np.mean(x1)
    v_x0 = np.var(x0)
    v_x1 = np.var(x1)
    w0 = (c_tx1 * c_x0x1 - v_x1 * c_tx0) / (c_x0x1**2 - v_x0 * v_x1)
    w1 = (c_tx0 * c_x0x1 - v_x0 * c_tx1) / (c_x0x1**2 - v_x0 * v_x1)
    w2 = -w0 * np.mean(x0) - w1 * np.mean(x1) + np.mean(t) 
    return np.array([w[0], w[1], w[2]])


# Sol1) 경사하강법 쓰지 않고(t에서 안빼줌)
def fit_plane2(x0,x1,t):   # ***오늘 쉅시간에 한 부분이 이부분! 판.서.사.진 참고!! ***
    X = np.ones([len(x0),3])  # len(x0) : 입력벡터의 개수만큼 행으로, 3 : 샘플의 차원(x0, x1, x2=1(바이어스))
    T = np.zeros([len(t),1])  # T : y벡터(판서에서). (y=w0x0 + w1x1 + w2의 y들 벡터)
    for i in range(len(x0)): # N-1까지 돎
        X[i,0] = x0[i] # i행의 1열  (행은 물론 0행이아니라 1행부터 시작^^)
        X[i,1] = x1[i] # i행의 2열
        T[i]=t[i] 
    X_T = np.transpose(X)  # x (행렬)에 T 시킴(역행렬 말고 트랜스포즈!)
    XTX_inv =np.linalg.inv(X_T.dot(X))  # 행렬 곱 후 역행렬 시키기
    XXX = XTX_inv.dot(X_T)  # 위에 역행렬 시킨거에다가 다시 행렬 곱
    w = XXX.dot(T) # T 행렬과 곱해주기 (T : y벡터(열벡터=행렬))
    return [w[0,0], w[1,0], w[2,0]]


# 평편의 기울기  
def dmse_plane(x0, x1, t, w):
    y = w[0] * x0 + w[1] * x1 + w[2]
    d_w0 = 2 * np.mean((y - t) * x0)
    d_w1 = 2 * np.mean((y - t) * x1)
    d_w2 = 2 * np.mean(y - t)
    return d_w0,d_w1,d_w2


# Sol2 ) 경사 하강법으로도 구할 수 있음!!
# 경사하강법  ------------------------------------ 
def fit_plane_num(x0,x1,t):
    w_init = [10.0, 10.0, 100.0] # 초기 매개 변수 
    alpha = 0.0001 # 학습률
    i_max = 100000 # 반복의 최대 수울기
    eps = 0.001 # 반복을 종료 시의 절대 값의 한계
    w = w_init
    for i in range(1, i_max):
        dmse = dmse_plane(x0, x1, t, w)
        w[0] = w[0] - alpha * dmse[0]
        w[1] = w[1] - alpha * dmse[1]
        w[2] = w[2] - alpha * dmse[2]
        if max(np.absolute(dmse)) < eps: #종료판정, np.absolute는 절대치 / 최적점에 도달하면 미분값=0에 수렴
            break
    return np.array([w[0], w[1], w[2]])

# 메인 ------------------------------------
plt.figure(figsize=(6, 5))
ax = plt.subplot(1, 1, 1, projection='3d')
W = fit_plane_num(X0, X1, T)
print("w0={0:.1f}, w1={1:.1f}, w2={2:.1f}".format(W[0], W[1], W[2]))
show_plane(ax, W)
show_data2(ax, X0, X1, T)
mse = mse_plane(X0, X1, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.show()
```

```python
# --- 리스트 5-2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# 데이터 로드 ----------------------------
outfile = np.load('ch5_data.npz')
X = outfile['X']
X_min = outfile['X_min']
X_max = outfile['X_max']
X_n = outfile['X_n']
T = outfile['T']
```

```python
# --- 리스트 5-2-(2)
# 가우스 함수 ---------------------------------
def gauss(x, mu, s):
    return np.exp(-(x - mu)**2 / (2 * s**2))
```

```python
# 리스트 5-2-(3)
# 메인 ------------------------------------
M = 4
plt.figure(figsize=(4, 4))
mu = np.linspace(5, 30, M)
s = mu[1] - mu[0] # (A)
xb = np.linspace(X_min, X_max, 100)
for j in range(M):
    y = gauss(xb, mu[j], s)
    plt.plot(xb, y, color='gray', linewidth=3)
plt.grid(True)
plt.xlim(X_min, X_max)
plt.ylim(0, 1.2)
plt.show()
```

```python
# 리스트 5-2-(4)
# 선형 기저 함수 모델 ----------------
def gauss_func(w, x):
    m = len(w) - 1
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    y = np.zeros_like(x) # x와 같은 크기로 요소가 0의 행렬 y를 작성
    for j in range(m):
        y = y + w[j] * gauss(x, mu[j], s)
    y = y + w[m]
    return y
```

```python
# 리스트 5-2-(5)
# 선형 기저 함수 모델 MSE ----------------
def mse_gauss_func(x, t, w):
    y = gauss_func(w, x)
    mse = np.mean((y - t)**2)
    return mse
```

```python
# 리스트 5-2-(6)
# 선형 기저 함수 모델 정확한 솔루션 -----------------
def fit_gauss_func(x, t, m):
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    n = x.shape[0]
    psi = np.ones((n, m+1))
    for j in range(m):
        psi[:, j] = gauss(x, mu[j], s)
    psi_T = np.transpose(psi)
    
    
    b = np.linalg.inv(psi_T.dot(psi))
    c = b.dot(psi_T)
    w = c.dot(t)
    return w
```

```python
# 리스트 5-2-(7)
# 가우스 기저 함수 표시 -----------------------
def show_gauss_func(w):
    xb = np.linspace(X_min, X_max, 100)
    y = gauss_func(w, xb)
    plt.plot(xb, y, c=[.5, .5, .5], lw=4) 


# 메인 ----------------------------------
plt.figure(figsize=(4, 4))
M = 4
W = fit_gauss_func(X, T, M)
show_gauss_func(W)
plt.plot(X, T, marker='o', linestyle='None',
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
mse = mse_gauss_func(X, T, W)
print('W='+ str(np.round(W,1)))
print("SD={0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```

```python
# 리스트 5-2-(12)
# K 분할 교차 검증 -----------------------------
def kfold_gauss_func(x, t, m, k):
    n = x.shape[0]
    mse_train = np.zeros(k)
    mse_test = np.zeros(k)
    for i in range(0, k):
        x_train = x[np.fmod(range(n), k) != i] # (A)
        t_train = t[np.fmod(range(n), k) != i] # (A)
        x_test = x[np.fmod(range(n), k) == i] # (A)
        t_test = t[np.fmod(range(n), k) == i] # (A)
        wm = fit_gauss_func(x_train, t_train, m)
        mse_train[i] = mse_gauss_func(x_train, t_train, wm)
        mse_test[i] = mse_gauss_func(x_test, t_test, wm)
    return mse_train, mse_test
```

```python
# 리스트 5-2-(17)
# 모델 A -----------------------------------
def model_A(x, w):
    y = w[0] - w[1] * np.exp(-w[2] * x)
    return y 


# 모델 A 표시 -------------------------------
def show_model_A(w):
    xb = np.linspace(X_min, X_max, 100)
    y = model_A(xb, w)
    plt.plot(xb, y, c=[.5, .5, .5], lw=4) 


# 모델 A의 MSE ------------------------------
def mse_model_A(w, x, t):
    y = model_A(x, w)
    mse = np.mean((y - t)**2)
    return mse
```

```python
# 리스트 5-2-(18)
from scipy.optimize import minimize 


# 모델 A의 매개 변수 최적화 -----------------
def fit_model_A(w_init, x, t):
    res1 = minimize(mse_model_A, w_init, args=(x, t), method="powell")
    return res1.x
```
